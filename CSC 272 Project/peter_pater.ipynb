Importing Packages

```python
import pandas as pd
import yfinance as yf
import datetime
import talib
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import random

```

Data Pre-Processing

```python
# Load data
start_date = '1800-01-01'
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')
symbol = "AAPL"
data = yf.download(symbol, start=start_date, end=end_date)

# Feature Engineering
data["Dollar_Volume"] = data["Adj Close"] * data["Volume"]
data["Pct_Change"] = data["Adj Close"].pct_change ()
data["Dollar_Volume_Pct_Change"] = data["Dollar_Volume"].pct_change()

# Replace infinite values with NaN and drop them
data.replace([np.inf, -np.inf], np.nan, inplace=True)


data["Pct_Change_Category"] = data["Pct_Change"].apply(lambda x: 'positive' if x > 0.01 else ('negative' if x < -0.01 else 'neutral'))

# Drop NA values after pct_change calculations
data = data.dropna()
data['Volume'] = data['Volume'].astype(float)

# data.reset_index(drop = True, inplace=True)

data.to_csv('apple_price.csv')
```

```python
from tqdm import tqdm
tqdm.pandas()
from scipy.stats import mode

def euclid_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn(test_row, train_data, k):
    return (train_data.apply(lambda x: euclid_distance(x, test_row)).sort_values().iloc[:k].index)

def cosine_similarity(x1, x2):
    return np.dot(x1, x2) / np.linalg.norm(x1) * np.linalg.norm(x2)

def knn_cosine(test_row, train_data, k):
    return (train_data.apply(lambda x: cosine_similarity(x, test_row)).sort_values().iloc[:k].index)

knned_euclid = data_test_series.progress_apply(lambda x: knn(x, data_train_series, 7))
new_knned_euclid = knned_euclid.apply(lambda x: mode(x)[0][0])

knned_cos = data_test_series.progress_apply(lambda x: knn_cosine(x, data_train_series, 7))
new_knned_cos = knned_cos.apply(lambda x: mode(x)[0][0])
```

Prediction

```python
# Evaluation Table
from tabulate import tabulate

models = [
    ("KNN (Euclidean) k = 3", y_test, y_pred_euclidean),
    ("KNN (Euclidean) k = 5", y_test, y_pred_euclidean),
    ("KNN (Euclidean) k = 7", y_test, y_pred_euclidean),
    ("KNN (Cosine) k = 3", y_test, y_pred_cosine),
    ("KNN (Cosine) k = 5", y_test, y_pred_cosine),
    ("KNN (Cosine) k = 7", y_test, y_pred_cosine),
    ("Majority Class", y_test, y_majority_pred),
    ("Random Classifier (Uniform)", y_test, y_random_pred_uniform),
    ("Random Classifier (Proportional)", y_test, y_random_pred_proportional)
]

results = []
for model_name, y_true, y_pred in models:
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    results.append([model_name, accuracy, precision, recall, f1])

headers = ["Model", "Accuracy", "Precision", "Recall", "F1-Score"]
print(tabulate(results, headers=headers, tablefmt="grid"))
```
